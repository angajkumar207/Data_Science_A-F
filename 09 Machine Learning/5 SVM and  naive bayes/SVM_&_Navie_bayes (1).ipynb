4. Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision
boundary:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

# Load the Iris dataset
iris = load_iris()
X = iris.data[:, :2]  # We take only the first two features (sepal length and sepal width) for easier visualization
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create an SVM Classifier with a Polynomial Kernel
poly_svm_classifier = SVC(kernel='poly', degree=3, gamma='auto', C=1.0, random_state=42)

# Train the classifier on the training data
poly_svm_classifier.fit(X_train, y_train)

# Evaluate the accuracy (optional, but good practice)
accuracy = poly_svm_classifier.score(X_test, y_test)
print(f"Accuracy of Polynomial SVM Classifier: {accuracy:.4f}")

# --- Visualize the decision boundary ---

# Create a meshgrid to plot the decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

# Predict the class for each point in the meshgrid
Z = poly_svm_classifier.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary and the data points
plt.figure(figsize=(10, 7))
plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.RdYlBu)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdYlBu, edgecolors='k', s=80)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('SVM Classifier with Polynomial Kernel Decision Boundary on Iris Dataset')
plt.show()
```

6. Write a Python program to train a Multinomial Naïve Bayes classifier for text classification using the 20
Newsgroups dataset.

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# Load the 20 Newsgroups dataset
# We'll select a few categories for a more focused example
categories = ['alt.atheism', 'soc.religion.christian', 'comp.graphics', 'sci.med']
newsgroups_train = fetch_20newsgroups(subset='train', categories=categories, shuffle=True, random_state=42)
newsgroups_test = fetch_20newsgroups(subset='test', categories=categories, shuffle=True, random_state=42)

print(f"Number of training samples: {len(newsgroups_train.data)}")
print(f"Number of testing samples: {len(newsgroups_test.data)}")
print(f"Categories: {newsgroups_train.target_names}")

# Vectorize the text data using TF-IDF
# Convert text documents to a matrix of TF-IDF features
vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(newsgroups_train.data)
X_test_tfidf = vectorizer.transform(newsgroups_test.data)

y_train = newsgroups_train.target
y_test = newsgroups_test.target

print(f"Shape of TF-IDF training data: {X_train_tfidf.shape}")
print(f"Shape of TF-IDF testing data: {X_test_tfidf.shape}")

# Create a Multinomial Naïve Bayes classifier
mnb_classifier = MultinomialNB()

# Train the classifier on the training data
mnb_classifier.fit(X_train_tfidf, y_train)

# Make predictions on the test set
y_pred = mnb_classifier.predict(X_test_tfidf)

# Evaluate the accuracy of the classifier
accuracy = accuracy_score(y_test, y_pred)
print(f"\nAccuracy of Multinomial Naïve Bayes Classifier: {accuracy:.4f}")

# Print classification report for more detailed evaluation
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=newsgroups_test.target_names))
```

7. Write a Python program to train an SVM Classifier with different C values and compare the decision
boundaries visually.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

# Load the Iris dataset
iris = load_iris()
X = iris.data[:, :2]  # Take only the first two features (sepal length and sepal width) for 2D visualization
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define different C values to compare
C_values = [0.1, 1, 10, 100]

# Create a mesh to plot the decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

plt.figure(figsize=(15, 10))

for i, C in enumerate(C_values):
    # Create an SVM Classifier with an RBF kernel and current C value
    svm_classifier = SVC(kernel='rbf', C=C, gamma='auto', random_state=42)

    # Train the classifier
    svm_classifier.fit(X_train, y_train)

    # Predict the class for each point in the mesh
    Z = svm_classifier.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot the decision boundary and the data points
    plt.subplot(2, 2, i + 1)
    plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.RdYlBu)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdYlBu, edgecolors='k', s=80)
    plt.xlabel('Sepal Length (cm)')
    plt.ylabel('Sepal Width (cm)')
    plt.title(f'SVM with C = {C} (Accuracy: {svm_classifier.score(X_test, y_test):.2f})')

plt.tight_layout()
plt.show()
```

8. Write a Python program to train a Bernoulli Naïve Bayes classifier for binary classification on a dataset with

```python
import numpy as np
from sklearn.naive_bayes import BernoulliNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# --- Create a synthetic binary dataset for demonstration ---
# Let's imagine features like:
# [is_email_spam, has_keyword_free, has_urgent_subject, has_attachment, is_from_unknown_sender]

# Generate 1000 samples
n_samples = 1000
n_features = 5

# Randomly generate binary features (0 or 1)
X_binary = np.random.randint(0, 2, size=(n_samples, n_features))

# Create a target variable (binary classification)
# For example, let's say spam is more likely if 'has_keyword_free' (feature 1) is 1,
# 'has_urgent_subject' (feature 2) is 1, and 'is_from_unknown_sender' (feature 4) is 1.
# Add some noise to make it realistic.
y_binary = ((X_binary[:, 1] == 1) | (X_binary[:, 2] == 1) | (X_binary[:, 4] == 1)).astype(int)
y_binary = np.logical_xor(y_binary, np.random.randint(0, 2, size=n_samples, high=3) == 0).astype(int) # Add some noise

print(f"Generated dataset with {n_samples} samples and {n_features} binary features.")
print(f"Example X (first 5 rows):\n{X_binary[:5]}")
print(f"Example y (first 5 labels): {y_binary[:5]}")

# Split the dataset into training and testing sets
X_train_binary, X_test_binary, y_train_binary, y_test_binary = train_test_split(X_binary, y_binary, test_size=0.3, random_state=42)

print(f"\nTraining set size: {X_train_binary.shape[0]}")
print(f"Testing set size: {X_test_binary.shape[0]}")

# Create a Bernoulli Naïve Bayes classifier
# binarize=None is important if features are already binary (0 or 1)
bnb_classifier = BernoulliNB(binarize=None)

# Train the classifier on the training data
bnb_classifier.fit(X_train_binary, y_train_binary)

print("\nBernoulli Naïve Bayes Classifier trained successfully.")

# Make predictions on the test set
y_pred_binary = bnb_classifier.predict(X_test_binary)

# Evaluate the accuracy of the classifier
accuracy_bnb = accuracy_score(y_test_binary, y_pred_binary)
print(f"\nAccuracy of Bernoulli Naïve Bayes Classifier: {accuracy_bnb:.4f}")
```

9. Write a Python program to apply feature scaling before training an SVM model and compare results with
unscaled data.

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# 1. Load a dataset (Iris dataset for demonstration)
iris = load_iris()
X = iris.data
y = iris.target

# 2. Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("--- SVM without Feature Scaling ---")
# 3. Train SVM on unscaled data
svm_unscaled = SVC(kernel='rbf', random_state=42)
svm_unscaled.fit(X_train, y_train)

# Make predictions and evaluate accuracy
y_pred_unscaled = svm_unscaled.predict(X_test)
accuracy_unscaled = accuracy_score(y_test, y_pred_unscaled)
print(f"Accuracy of SVM (unscaled data): {accuracy_unscaled:.4f}")

print("\n--- SVM with Feature Scaling ---")
# 4. Apply Feature Scaling (StandardScaler)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 5. Train SVM on scaled data
svm_scaled = SVC(kernel='rbf', random_state=42)
svm_scaled.fit(X_train_scaled, y_train)

# Make predictions and evaluate accuracy
y_pred_scaled = svm_scaled.predict(X_test_scaled)
accuracy_scaled = accuracy_score(y_test, y_pred_scaled)
print(f"Accuracy of SVM (scaled data): {accuracy_scaled:.4f}")

# 6. Compare results
print("\nComparison:")
print(f"Difference in accuracy (scaled - unscaled): {accuracy_scaled - accuracy_unscaled:.4f}")
if accuracy_scaled > accuracy_unscaled:
    print("Feature scaling improved the SVM's accuracy!")
elif accuracy_scaled < accuracy_unscaled:
    print("Feature scaling decreased the SVM's accuracy (this is rare for RBF kernel but can happen).")
else:
    print("Feature scaling had no significant impact on accuracy.")
```

10.  Write a Python program to train a Gaussian Naïve Bayes model and compare the predictions before and
after Laplace Smoothing.

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("--- Gaussian Naïve Bayes without (or minimal) var_smoothing ---")
# Create a Gaussian Naïve Bayes classifier with a very small var_smoothing
# Default var_smoothing is 1e-9. Let's use a smaller one to simulate 'less smoothing'.
gnb_no_smoothing = GaussianNB(var_smoothing=1e-10)

# Train the classifier
gnb_no_smoothing.fit(X_train, y_train)

# Make predictions and evaluate accuracy
y_pred_no_smoothing = gnb_no_smoothing.predict(X_test)
accuracy_no_smoothing = accuracy_score(y_test, y_pred_no_smoothing)
print(f"Accuracy of GNB (minimal var_smoothing): {accuracy_no_smoothing:.4f}")

print("\n--- Gaussian Naïve Bayes with increased var_smoothing (Laplace-like effect) ---")
# Create a Gaussian Naïve Bayes classifier with an increased var_smoothing
# A larger value can help prevent issues with zero variance in classes.
gnb_with_smoothing = GaussianNB(var_smoothing=1e-6) # Experiment with this value

# Train the classifier
gnb_with_smoothing.fit(X_train, y_train)

# Make predictions and evaluate accuracy
y_pred_with_smoothing = gnb_with_smoothing.predict(X_test)
accuracy_with_smoothing = accuracy_score(y_test, y_pred_with_smoothing)
print(f"Accuracy of GNB (with var_smoothing): {accuracy_with_smoothing:.4f}")

print("\n--- Comparison of Predictions ---")
print(f"First 10 predictions (minimal smoothing): {y_pred_no_smoothing[:10]}")
print(f"First 10 predictions (with smoothing):    {y_pred_with_smoothing[:10]}")
print(f"Actual first 10 labels:                   {y_test[:10]}")

print("\n--- Overall Accuracy Comparison ---")
print(f"Accuracy difference (with smoothing - minimal smoothing): {accuracy_with_smoothing - accuracy_no_smoothing:.4f}")

if accuracy_with_smoothing > accuracy_no_smoothing:
    print("Increased var_smoothing improved accuracy in this case.")
elif accuracy_with_smoothing < accuracy_no_smoothing:
    print("Increased var_smoothing slightly decreased accuracy in this case.")
else:
    print("Increased var_smoothing had no significant impact on accuracy in this case.")
```

11.  Write a Python program to train an SVM Classifier and use GridSearchCV to tune the hyperparameters (C,
gamma, kernel).

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define the SVM classifier
svm = SVC(random_state=42)

# Define the parameter grid to search
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [0.001, 0.01, 0.1, 1],
    'kernel': ['rbf', 'linear', 'poly']
}

# Create GridSearchCV object
# cv=5 means 5-fold cross-validation
# verbose=1 shows progress
# n_jobs=-1 uses all available processors
grid_search = GridSearchCV(svm, param_grid, cv=5, verbose=1, n_jobs=-1)

print("Starting GridSearchCV to find the best hyperparameters...")
# Fit GridSearchCV to the training data
grid_search.fit(X_train, y_train)

print("\nGridSearchCV completed.")

# Get the best parameters and best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

print(f"Best Parameters: {best_params}")
print(f"Best Cross-validation Accuracy: {best_score:.4f}")

# Get the best estimator (model) found by GridSearchCV
best_svm_model = grid_search.best_estimator_

# Make predictions on the test set with the best model
y_pred = best_svm_model.predict(X_test)

# Evaluate the accuracy of the best model on the test set
test_accuracy = accuracy_score(y_test, y_pred)
print(f"Test Accuracy of the best SVM model: {test_accuracy:.4f}")
```

12. Write a Python program to train an SVM Classifier on an imbalanced dataset and apply class weighting and
check it improve accuracy.

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# -----------------------------------------
# 1. Create an imbalanced dataset
# -----------------------------------------
X, y = make_classification(
    n_samples=5000,
    n_features=20,
    n_informative=2,
    n_redundant=10,
    n_clusters_per_class=1,
    weights=[0.95, 0.05],    # 95% of class 0, 5% of class 1
    random_state=42
)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# -----------------------------------------
# 2. SVM WITHOUT class weights
# -----------------------------------------
svm_default = make_pipeline(StandardScaler(), SVC(kernel="rbf"))
svm_default.fit(X_train, y_train)

y_pred_default = svm_default.predict(X_test)

print("----- SVM WITHOUT Class Weights -----")
print("Accuracy:", accuracy_score(y_test, y_pred_default))
print(classification_report(y_test, y_pred_default))


# -----------------------------------------
# 3. SVM WITH class weights
# -----------------------------------------
svm_weighted = make_pipeline(
    StandardScaler(),
    SVC(kernel="rbf", class_weight="balanced")
)
svm_weighted.fit(X_train, y_train)

y_pred_weighted = svm_weighted.predict(X_test)

print("----- SVM WITH Class Weights (balanced) -----")
print("Accuracy:", accuracy_score(y_test, y_pred_weighted))
print(classification_report(y_test, y_pred_weighted))
```

13. Write a Python program to implement a Naïve Bayes classifier for spam detection using email data.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# -----------------------------------------------------------
# 1. Sample email spam dataset
#    (Replace "emails.csv" with your real dataset)
# -----------------------------------------------------------

# Expected CSV format:
# text,label
# "Hello, how are you?", "ham"
# "Win money now!!!", "spam"

data = pd.read_csv("emails.csv")

# Convert labels to binary format (spam = 1, ham = 0)
data['label'] = data['label'].map({'ham': 0, 'spam': 1})

X = data['text']
y = data['label']

# -----------------------------------------------------------
# 2. Split into train and test sets
# -----------------------------------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

# -----------------------------------------------------------
# 3. Convert email text into TF-IDF features
# -----------------------------------------------------------
vectorizer = TfidfVectorizer(stop_words='english')
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# -----------------------------------------------------------
# 4. Train Naïve Bayes classifier
# -----------------------------------------------------------
model = MultinomialNB()
model.fit(X_train_tfidf, y_train)

# -----------------------------------------------------------
# 5. Predict on test set
# -----------------------------------------------------------
y_pred = model.predict(X_test_tfidf)

# -----------------------------------------------------------
# 6. Evaluate results
# -----------------------------------------------------------
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

14. Write a Python program to train an SVM Classifier and a Naïve Bayes Classifier on the same dataset and
compare their accuracy.

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

# -----------------------------------------------------
# 1. Load dataset (using IRIS for demonstration)
# -----------------------------------------------------
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# -----------------------------------------------------
# 2. Train SVM Classifier
# -----------------------------------------------------
svm_model = SVC(kernel='rbf')
svm_model.fit(X_train, y_train)

svm_pred = svm_model.predict(X_test)

print("===== SVM Classifier Results =====")
print("Accuracy:", accuracy_score(y_test, svm_pred))
print(classification_report(y_test, svm_pred))

# -----------------------------------------------------
# 3. Train Naïve Bayes Classifier
# -----------------------------------------------------
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)

nb_pred = nb_model.predict(X_test)

print("===== Naïve Bayes Classifier Results =====")
print("Accuracy:", accuracy_score(y_test, nb_pred))
print(classification_report(y_test, nb_pred))

# -----------------------------------------------------
# 4. Compare Accuracies
# -----------------------------------------------------
print("===== Accuracy Comparison =====")
print(f"SVM Accuracy: {accuracy_score(y_test, svm_pred):.4f}")
print(f"Naïve Bayes Accuracy: {accuracy_score(y_test, nb_pred):.4f}")
```

15. Write a Python program to train an SVM Classifier using One-vs-Rest (OvR) and One-vs-One (OvO)
strategies on the Wine dataset and compare their accuracy.

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.multiclass import OneVsRestClassifier, OneVsOneClassifier
from sklearn.metrics import accuracy_score, classification_report

# ------------------------------------------------------
# 1. Load Wine dataset
# ------------------------------------------------------
data = load_wine()
X = data.data
y = data.target

# Split data (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Scale features for SVM
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# ------------------------------------------------------
# 2. Train SVM using One-vs-Rest (OvR)
# ------------------------------------------------------
ovr_model = OneVsRestClassifier(SVC(kernel='rbf'))
ovr_model.fit(X_train, y_train)

ovr_pred = ovr_model.predict(X_test)

print("===== One-vs-Rest (OvR) Results =====")
print("Accuracy:", accuracy_score(y_test, ovr_pred))
print(classification_report(y_test, ovr_pred))

# ------------------------------------------------------
# 3. Train SVM using One-vs-One (OvO)
# ------------------------------------------------------
ovo_model = OneVsOneClassifier(SVC(kernel='rbf'))
ovo_model.fit(X_train, y_train)

ovo_pred = ovo_model.predict(X_test)

print("===== One-vs-One (OvO) Results =====")
print("Accuracy:", accuracy_score(y_test, ovo_pred))
print(classification_report(y_test, ovo_pred))

# ------------------------------------------------------
# 4. Comparison
# ------------------------------------------------------
print("===== Accuracy Comparison =====")
print(f"OvR Accuracy: {accuracy_score(y_test, ovr_pred):.4f}")
print(f"OvO Accuracy: {accuracy_score(y_test, ovo_pred):.4f}")
```

17.  Write a Python program to train an SVM Classifier using Stratified K-Fold Cross-Validation and compute the
average accuracy

```python
from sklearn.datasets import load_breast_cancer
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import numpy as np

# ----------------------------------------------------
# 1. Load Dataset
# ----------------------------------------------------
data = load_breast_cancer()
X = data.data
y = data.target

# Standardize data (important for SVM)
scaler = StandardScaler()
X = scaler.fit_transform(X)

# ----------------------------------------------------
# 2. Create Stratified K-Fold
# ----------------------------------------------------
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

accuracies = []

# ----------------------------------------------------
# 3. Train SVM on each fold
# ----------------------------------------------------
for train_index, test_index in kfold.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    model = SVC(kernel='rbf')
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)

    print(f"Fold Accuracy: {acc:.4f}")

# ----------------------------------------------------
# 4. Average Accuracy
# ----------------------------------------------------
avg_accuracy = np.mean(accuracies)
print("\n===============================")
print(f"Average Accuracy: {avg_accuracy:.4f}")
print("===============================")
```

18. Write a Python program to train a Naïve Bayes classifier using different prior probabilities and compare
performance.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

# ----------------------------------------------------
# 1. Load dataset
# ----------------------------------------------------
data = load_iris()
X = data.data
y = data.target

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# ----------------------------------------------------
# 2. Define different prior probability settings
# ----------------------------------------------------
priors_list = [
    None,                                    # Let the model learn from data
    [0.6, 0.3, 0.1],                          # Custom priors
    [0.33, 0.33, 0.34],                       # Nearly equal priors
]

# ----------------------------------------------------
# 3. Train and evaluate for each prior setting
# ----------------------------------------------------
for idx, priors in enumerate(priors_list):
    print(f"\n========== Model {idx+1} (Priors = {priors}) ==========")

    model = GaussianNB(priors=priors)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    print("Accuracy:", accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred))
```

23.  Write a Python program to train an SVM Classifier and visualize the Confusion Matrix using seaborn.

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix

# ---------------------------------------------------------
# 1. Load Dataset
# ---------------------------------------------------------
data = load_breast_cancer()
X = data.data
y = data.target

# Split training/testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Scale features (important for SVM)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# ---------------------------------------------------------
# 2. Train SVM Classifier
# ---------------------------------------------------------
model = SVC(kernel='rbf')
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# ---------------------------------------------------------
# 3. Compute Confusion Matrix
# ---------------------------------------------------------
cm = confusion_matrix(y_test, y_pred)

# ---------------------------------------------------------
# 4. Visualize with seaborn
# ---------------------------------------------------------
plt.figure(figsize=(6, 4))
sns.heatmap(
    cm, annot=True, fmt="d", cmap="Blues",
    xticklabels=["Benign (0)", "Malignant (1)"],
    yticklabels=["Benign (0)", "Malignant (1)"]
)
plt.title("SVM Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.tight_layout()
plt.show()
```

```python
24.  Write a Python program to train an SVM Regressor (SVR) and evaluate its performance using Mean Absolute
Error (MAE) instead of MSE.
```

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error

# ---------------------------------------------------------
# 1. Load Dataset
# ---------------------------------------------------------
data = fetch_california_housing(as_frame=False)
X = data.data
y = data.target

# Split training/testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Scale features (very important for SVR)
scaler_X = StandardScaler()
scaler_y = StandardScaler()

X_train_scaled = scaler_X.fit_transform(X_train)
X_test_scaled = scaler_X.transform(X_test)

# Reshape y for scaling
y_train_scaled = scaler_y.fit_transform(y_train.reshape(-1, 1)).ravel()

# ---------------------------------------------------------
# 2. Train SVM Regressor (SVR)
# ---------------------------------------------------------
model = SVR(kernel='rbf')
model.fit(X_train_scaled, y_train_scaled)

# Predict (scaled output)
y_pred_scaled = model.predict(X_test_scaled)

# Inverse scaling to original units
y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1)).ravel()

# ---------------------------------------------------------
# 3. Evaluate Using Mean Absolute Error (MAE)
# ---------------------------------------------------------
mae = mean_absolute_error(y_test, y_pred)

print("=========== SVR Regression Performance ===========")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
```

25. Write a Python program to train a Naïve Bayes classifier and evaluate its performance using the ROC-AUC
score.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score, roc_curve

# ---------------------------------------------------------
# 1. Load Dataset
# ---------------------------------------------------------
data = load_breast_cancer()
X = data.data
y = data.target   # Binary labels (0/1) → suitable for ROC-AUC

# Split into training/testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Standardize features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# ---------------------------------------------------------
# 2. Train Naïve Bayes Classifier
# ---------------------------------------------------------
model = GaussianNB()
model.fit(X_train, y_train)

# Predict probability for the positive class
y_prob = model.predict_proba(X_test)[:, 1]

# ---------------------------------------------------------
# 3. Compute ROC-AUC Score
# ---------------------------------------------------------
auc = roc_auc_score(y_test, y_prob)
print("========== Naïve Bayes ROC-AUC ==========")
print(f"ROC-AUC Score: {auc:.4f}")

# ---------------------------------------------------------
# 4. Plot ROC Curve
# ---------------------------------------------------------
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, label=f"AUC = {auc:.4f}")
plt.plot([0, 1], [0, 1], 'k--')  # Random baseline
plt.title("Naïve Bayes ROC Curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate (Recall)")
plt.legend()
plt.tight_layout()
plt.show()
```

26. Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import precision_recall_curve

# Load dataset
data = load_breast_cancer()
X = data.data
y = data.target

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Feature scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Train SVM with probability estimates enabled
svm_model = SVC(kernel='rbf', probability=True)
svm_model.fit(X_train, y_train)

# Get predicted probabilities
y_scores = svm_model.predict_proba(X_test)[:, 1]

# Compute Precision-Recall values
precision, recall, thresholds = precision_recall_curve(y_test, y_scores)

# Plot Precision-Recall Curve
plt.figure(figsize=(7, 5))
plt.plot(recall, precision, linewidth=2)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve for SVM Classifier")
plt.grid(True)
plt.tight_layout()
plt.show()
```
